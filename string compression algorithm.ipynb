## Directions

The problem at hand is to implement the following compression algorithm:

- The input is a string, and the output is a compressed string.

- A valid input consists of zero or more upper case english letters A-Z.

- To produce the compressed output, any run of two or more of the same
character should be converted to two of that character plus a number
indicating how many repeated runs were compressed. 

Examples:

    A --> A
    AA --> AA0
    AAA --> AA1
    AAAA --> AA2
    
    
- Only one digit may be used at a time, so if the run is quite long,
then you must use multiple character/number pairs


    AAAAAAAAAA --> AA8
    AAAAAAAAAAA --> AA9
    AAAAAAAAAAAA --> AA9A
    AAAAAAAAAAAAA --> AA9AA0
    AAAAAAAAAAAAAA --> AA9AA1


The decompression algorithm simply reverses this process. Please
implement both compress and decompress.

(1) The code, written in Java, Scala, Python, Ruby, Go, or JavaScript

(2) Instructions on how to run the code

(3) Unit tests that demonstrate the correctness of your two functions (compress 
and decompress) For example, here are a few interesting inputs that you could 
test with:


    A --> A
    ABAC --> ABAC
    AA --> AA0
    AAB --> AA0B
    AABB --> AA0BB0
    AAACBBC --> AA1CBB0C


```python
def compression(input_s=''):
    if not compressible_input(input_s):
        return 'The input is invalid'
    s = str(input_s) + '&'  # '&' added as a delimeter
    l = len(input_s)
    result = ''
    index_offset = 0

    if l <= 1:
        return input_s
    else:
        for i in range(l):
            if s[i] != s[i + 1]:
                reps = i + 1 - index_offset
                index_offset = i + 1
                result += compress_format(s[i], reps)
    return result


def compress_format(letter, reps):
    """takes a 1-char string and an int, emits a string
    to convert the char into the compressed format"""
    if reps == 1:
        return letter
    elif reps > 12:
        return letter * 2 + '9' + compress_format(letter, reps - 11)
    else:
        return letter * 2 + str(reps - 2)


def compressible_input(input_s):
    """A valid input consists of zero or more upper case english letters A-Z"""
    try:
        return (input_s.isalpha() and input_s.isupper() or input_s == '')
    except:
        return False
```

```python
def test_compression_valid():
    assert compression('') == ''
    assert compression() == ''
    assert compression('A') == 'A'
    assert compression('AAA') == 'AA1'
    assert compression('ABAC') == 'ABAC'
    assert compression('AA') == 'AA0'
    assert compression('AAB') == 'AA0B'
    assert compression('AABB') == 'AA0BB0'
    assert compression('AAACBBC') == 'AA1CBB0C'
    assert compression('KAAACBBCHH') == 'KAA1CBB0CHH0'
    assert compression('AAAAAAAAAAAAAA') == 'AA9AA1'
    assert compression('AAAAAAAAAAAAAABBBBBBBBBBBBBBB') == 'AA9AA1BB9BB2'


def test_compression_invalid():
    assert compression('88') == 'The input is invalid'
    assert compression('AA88') == 'The input is invalid'
    assert compression('aa') == 'The input is invalid'
    assert compression('BBaa') == 'The input is invalid'
    assert compression('BB0') == 'The input is invalid'
    assert compression(99) == 'The input is invalid'
    assert compression('&') == 'The input is invalid'
    assert compression({'j': 4, 'h': 5}) == 'The input is invalid'
    
test_compression_invalid()
test_compression_valid()

print("Tests pass")
    
```

```python
def decompression(input_s=''):
    if not decompressible_input(input_s):
        return 'The input is invalid'
    s = str(input_s) + '&'  # '&' added as a delimeter
    l = len(input_s)
    result = ''
    prev_letter = ''
    reps = 0

    if l <= 1:
        return input_s
    else:
        for i in range(l):
            if s[i].isdigit() and s[i + 1].isdigit():
                # two digits in a row is invalid
                return 'The input is invalid'
            elif s[i].isdigit():
                if not (s[i - 1] == s[i - 2] == prev_letter):
                    # invalid if the last two chars before digit don't match
                    return 'The input is invalid'
                reps = int(s[i])
                result += decompress_format(prev_letter, reps)
                reps = 0
            elif i > 1 and s[i - 2: i + 1] == s[i] * 3:
                # invalid if the there are 3 matching chars in a row
                return 'The input is invalid'
            else:
                result += s[i]
            prev_letter = s[i]
    return result


def decompress_format(letter, reps):
    """takes a 1-char string and an int, emits a string
    to convert the char into the decompressed format"""
    if reps == 0:
        return ""
    else:
        return (letter * reps)


def decompressible_input(input_s):
    """A valid input consists of zero or more
    upper case english letters A-Z and numers 0-9
    input must begin with a letter"""
    try:
        return input_s == '' or (input_s[0].isalpha() and input_s.isupper() and input_s.isalnum())
    except:
        return False
    
```
